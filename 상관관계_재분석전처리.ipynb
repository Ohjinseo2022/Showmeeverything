# 2012년

```python
# 2012 년 기준 아파트정보 변수 초기 추출
# 단지명으로 브랜드 지수 추출 예정 
apt2012 = apt[apt['년']==2012][['행정구','단지명','전용면적(㎡)','평단가','층','준공년도']]

apt2012_list = apt2012.values.tolist()

#사업체수 행만 추출하여 리스트로 전환처리
company = company.loc[company['항목']=='사업체수']
company_list = company.values.tolist()
#지하철개수
subway_list = subway.values.tolist()

## 유통업체개수 
mart_list = mart.values.tolist()
#인프라정보 
infra_list = infra.values.tolist()

#자동차 등록 대수 

car_list = car.values.tolist()

# 0번째핼 제거 
#hospital = hospital.drop([0])
#병원개수 리스로 전환 
hospital_list = hospital.values.tolist()

for i in range(len(apt2012_list)):
  apt2012_list[i][5] = int(2012-apt2012_list[i][5])
  for j in range(len(hospital_list)):
    if apt2012_list[i][0] == hospital_list[j][0]:
      #년도별 병원갯수
      apt2012_list[i].append(int(hospital_list[j][1]))
  for j in range(len(company_list)):
    if apt2012_list[i][0] == company_list[j][1]:
      #년도별 사업체수
      apt2012_list[i].append(company_list[j][4])
  for j in range(len(subway_list)):    
    if apt2012_list[i][0] == subway_list[j][0]:
      apt2012_list[i].append(subway_list[j][1])
  for j in range(len(school_list)):
    if apt2012_list[i][0] == school_list[j][0]:
      apt2012_list[i].append(school_list[j][1])
  for j in range(len(brand_list)):
    #브랜드 지수 비교 처리
    if brand_list[j][0] in apt2012_list[i][1]:
      apt2012_list[i].append(brand_list[j][1])
  # 브랜드 아파트에 속하지 않을경우 일괄적인 값을 넣어주기 위한코드
  if len(apt2012_list[i]) < 11:
      apt2012_list[i].append(50)
  ## 이름이 중복 된경우 ex) 래미안e편한세상 ㄷㄷ...
  elif len(apt2012_list[i]) == 12:
       apt2012_list[i].pop(-1)
  for j in range(len(car_list)):
    if apt2012_list[i][0] == car_list[j][0]:
      apt2012_list[i].append(int(car_list[j][7].replace(',','')))
  for j in range(len(infra_list)):
    if apt2012_list[i][0] == infra_list[j][0]:
      apt2012_list[i].append(int(infra_list[j][1]))
  for j in range(len(mart_list)):
    if apt2012_list[i][0] == mart_list[j][0]:
      apt2012_list[i].append(int(mart_list[j][1]))
  for j in range(len(infra_list)):
    if apt2012_list[i][0] == infra_list[j][0]:
      apt2012_list[i].append(int(infra_list[j][4].replace(',','')))
  
col_name=['행정구' , '단지' ,'전용면적','평단가' , '층수' ,'아파트연식', '병원수' , '사업체수', '지하철수' , '학군점수' , '브랜드지수','자동차등록수','학생1만명당학원수','유통업체수','문화생활비']
apt2012df = pd.DataFrame(apt2012_list,columns=col_name)
apt2012df = apt2012df[['행정구','전용면적','층수','아파트연식','병원수','사업체수','지하철수','학군점수','브랜드지수','자동차등록수','학생1만명당학원수','유통업체수','문화생활비','평단가']]
apt2012df.to_csv('/content/drive/MyDrive/시각화프로젝트/상관계수분석/2012년.csv',encoding='cp949',index=None)
```
