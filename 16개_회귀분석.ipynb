```python
apt = apt.sort_values(['평단가'])
apt_temp = pd.DataFrame({})
for i in tqdm(range(len(apt)//80)):
  if i == 0 :
    apt_temp= pd.concat([apt_temp,apt.iloc[0:(i+1)*80,1:].mean()],axis=1)
  else:
    apt_temp= pd.concat([apt_temp,apt.iloc[i*80:(i+1)*80,1:].mean()],axis=1)
apt_temp= pd.concat([apt_temp,apt.iloc[len(apt)-len(apt)%80:,1:].mean()],axis=1)
apt_temp = apt_temp.transpose()
## 인덱스번호 재설정
apt_temp = apt_temp.reset_index(drop=True)
apt_temp.to_csv('/content/drive/MyDrive/시각화프로젝트/2020데이터평균치계산.csv',encoding='cp949',index=None)
```

```python
from sklearn.model_selection import train_test_split ## 학습데이터와 테스트 데이터를 구분하기위한 라이브러리 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error ,mean_squared_error
```

# 데이터 정규화 처리

```python
scaler = MinMaxScaler()

df_scaled= apt_temp.iloc[:,:-1] # 마지막열임을 나타내는 -1은 포함하지 않음 -> 평단가 데이터를 제외한 데이터의 영향력차이를 줄여주기위함 
scaler.fit(df_scaled)
df_scaled=scaler.transform(df_scaled)

# 스케일린된 변환된 값을 데이터 프레임에 반영 
apt_temp.iloc[:,:-1] = df_scaled[:,:]
apt_temp = apt_temp.sort_values(by='평단가' ,ascending=True)

apt_temp
```

```python
### x_col 의 컬럼으로 학습모델 선정후 학습데이터와 테스트 데이터 분리
x_col = ['학생1만명당학원수',"노령화지수","지역종사자수",'사업체수','지하철수','병원수','자동차등록수','학군점수','유통업체수','층수','브랜드지수',"버스정류장수"]
x_df = apt_temp[x_col]
y_df = apt_temp['평단가']

x_train, x_test, y_train, y_test = train_test_split(x_df,y_df,test_size= 0.3,shuffle=True,random_state=12)

print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

## 2020년 정보 선형회귀모델 선정후 머신러닝 학습

```python
###Linear Regression 다중선형회귀
model_lr = LinearRegression()
model_lr.fit(x_train,y_train)
import joblib
# model save
joblib.dump(model_lr,'/content/drive/MyDrive/머신러닝모델저장소/apt2020_model.pkl')
print("회귀계수(기울기",np.round(model_lr.coef_,1))
print('상수항(절편)',np.round(model_lr.intercept_,1))
```

```python
y_lr_pred = model_lr.predict(x_test)

y_train_pred=model_lr.predict(x_train)

train_mae=mean_absolute_error(y_train, y_train_pred) #훈련 데이터의 평가 점수
print('평균 절대 오차')
print("Train MAE:%.4f" % train_mae)

test_mae=mean_absolute_error(y_test, y_lr_pred)
print("Test MAE:%.4f" % test_mae)
```

```python
print('평균제곱오차')
train_mae=mean_squared_error(y_train, y_train_pred) #훈련 데이터의 평가 점수
print("Train MSE:%.4f" % train_mae)

test_mae=mean_squared_error(y_test, y_lr_pred)
print("Test MSE:%.4f" % test_mae)
```

```python
print('평균 제곱근 편차')
train_mae=mean_squared_error(y_train, y_train_pred) #훈련 데이터의 평가 점수
print("Train RMSE:%.4f" % np.sqrt(train_mae))

test_mae=mean_squared_error(y_test, y_lr_pred)
print("Test RMSE:%.4f" % np.sqrt(test_mae))
```
